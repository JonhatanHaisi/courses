# Study Case 1 - Brazilian Population

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_data = np.array([])
y_data = np.array([])

with open("./populacao-brasileira.csv", "r") as file:
    for line in file.readlines()[1:]:
        year, population = line.split(";")
        x_data = np.append(x_data, int(year))
        y_data = np.append(y_data, int(population))
```

```python
plt.title("Brazilian population growth from 1980 to 2016")
plt.ylabel("Population X 100,000,000")
plt.xlabel("Year")

plt.bar(x_data, y_data, color="#E4E4E4")
plt.plot(x_data, y_data, color="k", linestyle="--")
plt.savefig("population.png", dpi=300)
```
