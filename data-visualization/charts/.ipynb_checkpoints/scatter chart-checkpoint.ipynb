# Scatter Chart

```python
import matplotlib.pyplot as plt
```

```python
title = "Line Chart"
x_label = "X Axis"
y_label = "Y Axis"

x_data = [1, 2, 3, 4, 5, 6]
y_data = [9, 6, 7, 3, 1, 4]
```

```python
plt.title(title)
plt.xlabel(x_label)
plt.ylabel(y_label)

plt.scatter(x_data, y_data)
```

```python
z_data = [200, 1000, 80, 500, 800, 100]
```

```python
plt.title(title)
plt.xlabel(x_label)
plt.ylabel(y_label)

plt.scatter(x_data, y_data, z_data)
```
