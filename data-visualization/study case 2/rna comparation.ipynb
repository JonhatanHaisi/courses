# Study Case 2: RNA Comparation

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("./human-rna.txt", "r") as human:
    human_rna = human.read().replace("\n", "")

with open("./bacterium-rna.txt", "r") as bacterium:
    bacterium_rna = bacterium.read().replace("\n", "")
```

```python
amino_acids = [ "A", "G", "T", "C"  ]
combinations = [ a+b for a in amino_acids for b in amino_acids ]
```

```python
human_count = { combination:0 for combination in combinations }
bacterium_count = { combination:0 for combination in combinations }
```

```python
human_pairs = [ human_rna[i]+human_rna[i+1]  for i in range(0, len(human_rna)-1, 2) ]
bacterium_pairs = [ bacterium_rna[i]+bacterium_rna[i+1]  for i in range(0, len(bacterium_rna)-1, 2) ]
```

```python
for pair in human_pairs:
    if pair in human_count:
        human_count[pair]=human_count[pair]+1
    
for pair in bacterium_pairs:
    if pair in bacterium_count:
        bacterium_count[pair]=bacterium_count[pair]+1
```

```python
fig, ax = plt.subplots()
im = ax.imshow(np.array([ human_count[key] for key in human_count]).reshape((4, 4)))

plt.colorbar(im)

# Loop over data dimensions and create text annotations.
for i in range(4):
    for j in range(4):
        text = ax.text(j, i, combinations[i + j],
                       ha="center", va="center", color="w")

ax.set_title("Human RNA")
```

```python
fig, ax = plt.subplots()
im = ax.imshow(np.array([ bacterium_count[key] for key in bacterium_count]).reshape((4, 4)))

plt.colorbar(im)

# Loop over data dimensions and create text annotations.
for i in range(4):
    for j in range(4):
        text = ax.text(j, i, combinations[i + j],
                       ha="center", va="center", color="w")

ax.set_title("Bacterium RNA")
```
